# 1. 유클리드 거리

```python
import numpy as np
```

- 두 점 사이의 '최단 직선 거리'를 재는 방식
- 비유 : A와 B 사이의 실제 거리를 자로 재는 것과 같다
- 특징 : 거리가 0에 가까울수록 두 문서가 비슷하다. (코사인 유사도는 1에 가까워야 비슷하지만, 여기선 숫자가 작아야 한다)


```python
def dist(x,y):
    return np.sqrt(np.sum((x-y)**2))

doc1 = np.array((2,3,0,1))
doc2 = np.array((1,2,3,1))
doc3 = np.array((2,1,2,2))
docQ = np.array((1,1,0,1))

print('문서1과 문서Q의 거리 :',dist(doc1,docQ))
print('문서2과 문서Q의 거리 :',dist(doc2,docQ))
print('문서3과 문서Q의 거리 :',dist(doc3,docQ))
```

```python
# 거리를 계산하는 함수
def dist(x,y):
    return np.sqrt(np.sum((x-y)**2))

# 예시 데이터 (단어 등장 횟수)
# [사과, 바나나, 포도, 오렌지] 순서대로 나온 횟수

doc1 = np.array((5, 0, 0, 0)) # 사과만 5개 나온 문서
doc2 = np.array((4, 1, 0, 0)) # 사과 4개, 바나나 1개 나온 문서
doc3 = np.array((0, 0, 5, 0)) # 포도만 5개 나온 문서 (기준점)

print('문서 1과 문서3의 거리 :', dist(doc1, doc3))
print('문서 2과 문서3의 거리 :', dist(doc2, doc3))
```

# 2. 자카드 유사도

- 공통으로 등장한 단어의 비율
- 비유 : 두 친구가 가진 '좋아하는 과일 목록' 비교할 때, 두 명 다 좋아하는 과일이 얼마나 많은지 비율로 계산하기
- 계산 $\frac{\text{교집합(공통 단어)}}{\text{합집합(전체 단어)}}$
- 특징 : 수치가 1에 가까울 수록 비슷하다

```python
doc1 = "apple banana everyone like likey watch card holder"
doc2 = "apple banana coupon passport love you"

tokenized_doc1 = doc1.split()
tokenized_doc2 = doc2.split()

print('문서1 :',tokenized_doc1)
print('문서2 :',tokenized_doc2)
```

```python
union = set(tokenized_doc1).union(set(tokenized_doc2))
print('문서1과 문서2의 합집합 :',union)
```

```python
intersection = set(tokenized_doc1).intersection(set(tokenized_doc2))
print('문서1과 문서2의 교집합 :',intersection)
```

```python
print('자카드 유사도 :',len(intersection)/len(union))
```

```python
# 1단계 : 문장 입력
doc4 = "딥러닝 실습은 정말 어렵지만 재미있다"
doc5 = "딥러닝 공부는 정말 힘들지만 재미있다"

# 2단계 : 단어 단위로 쪼개보기 (Tokenization)
tokenized_doc4 = doc4.split()
tokenized_doc5 = doc5.split()

# 3단계 : 합집합(Union)과 교집합(Intersection)구하기
union = set(tokenized_doc4).union(set(tokenized_doc5))
intersection = set(tokenized_doc4).intersection(set(tokenized_doc5))

# 4단계 : 유사도 계산 (교집합 개수 / 합집합 개수)
print('자카드 유사도 :',len(intersection)/len(union))
```
