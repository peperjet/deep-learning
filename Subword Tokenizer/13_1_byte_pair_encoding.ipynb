이 자료는 위키독스 딥 러닝을 이용한 자연어 처리 입문의 Byte Pair Encoding의 튜토리얼 자료입니다.

```python
import re, collections
from IPython.display import display, Markdown, Latex
```

- re : 파이썬의 정규 표현식 모듈. 문자열에서 특정 패턴을 검색, 대체, 분리 하는 작업할 때 사용
- collections : 정리함



```python
num_merges = 10
```

```python
dictionary = {'l o w </w>' : 5,
              'l o w e r </w>' : 2,
              'n e w e s t </w>' : 6,
              'w i d e s t </w>' : 3
              }
```

```python
def get_stats(dictionary):
    # 유니그램의 pair들의 빈도수를 카운트
    pairs = collections.defaultdict(int)
    for word, freq in dictionary.items():
        symbols = word.split()
        for i in range(len(symbols)-1):
            pairs[symbols[i],symbols[i+1]] += freq
    print('현재 pair들의 빈도수 :', dict(pairs))
    return pairs

def merge_dictionary(pair, v_in):
    v_out = {}
    bigram = re.escape(' '.join(pair))
    p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
    for word in v_in:
        w_out = p.sub(''.join(pair), word)
        v_out[w_out] = v_in[word]
    return v_out

bpe_codes = {}
bpe_codes_reverse = {}

for i in range(num_merges):
    display(Markdown("### Iteration {}".format(i + 1)))
    pairs = get_stats(dictionary)
    best = max(pairs, key=pairs.get)
    dictionary = merge_dictionary(best, dictionary)

    bpe_codes[best] = i
    bpe_codes_reverse[best[0] + best[1]] = best

    print("new merge: {}".format(best))
    print("dictionary: {}".format(dictionary))
```

bpe_codes를 출력하면 merge 했던 기록이 출력됩니다.

OOV에 대처하기

```python
def get_pairs(word):
    """Return set of symbol pairs in a word.
    Word is represented as a tuple of symbols (symbols being variable-length strings).
    """

    pairs = set()
    prev_char = word[0]
    for char in word[1:]:
        pairs.add((prev_char, char))
        prev_char = char
    return pairs


def encode(orig):
    """Encode word based on list of BPE merge operations, which are applied consecutively"""

    word = tuple(orig) + ('</w>',)
    display(Markdown("__word split into characters:__ <tt>{}</tt>".format(word)))

    pairs = get_pairs(word)

    if not pairs:
        return orig

    iteration = 0

    while True:
        iteration += 1
        display(Markdown("__Iteration {}:__".format(iteration)))

        print("bigrams in the word: {}".format(pairs))
        best = min(pairs, key = lambda pair: bpe_codes.get(pair, float('inf')))
        print("candidate for merging: {}".format(best))
        if best not in bpe_codes:
            display(Markdown("__Candidate not in BPE merges, algorithm stops.__"))
            break

        first, second = best
        new_word_list = []
        i = 0
        while i < len(word):
            try:
                j = word.index(first, i)
                new_word_list.extend(word[i:j])
                i = j
            except ValueError: # If first is not found in the remaining part of the word
                new_word_list.extend(word[i:])
                break

            # Now, at index i, word[i] is 'first'. Check if 'second' immediately follows 'first'.
            if i < len(word) - 1 and word[i] == first and word[i+1] == second:
                new_word_list.append(first + second)
                i += 2
            else: # If 'second' doesn't follow or it's the last element
                new_word_list.append(word[i])
                i += 1

        word = tuple(new_word_list) # Convert list back to tuple for next iteration
        print("word after merging: {}".format(word))
        if len(word) == 1:
            break
        else:
            pairs = get_pairs(word)

    # 특별 토큰인 </w>는 출력하지 않는다.
    if word and word[-1] == '</w>':
        word = word[:-1]
    elif word and word[-1].endswith('</w>'):
        word = word[:-1] + (word[-1].replace('</w>',''),)

    return word
```

encode 함수는 다음 단계를 통해 단어를 BPE로 인코딩합니다:

  - 입력 단어에 </w> 토큰을 추가하여 처리할 준비를 합니다.
  - 현재 단어에서 모든 인접한 문자 쌍을 추출합니다.
  - bpe_codes에 저장된 병합 규칙을 바탕으로 가장 우선순위가 높은 쌍을 찾아 병합합니다.
  - 더 이상 병합할 쌍이 없거나 단어가 하나의 토큰이 될 때까지 이 과정을 반복합니다.
  - 마지막으로 </w> 토큰을 제거하고 최종 BPE 인코딩 결과를 반환합니다.
