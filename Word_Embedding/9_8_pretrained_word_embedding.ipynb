1. Kearas Embedding layer 케라스 임베딩 레이어


```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
sentences = ['nice great best amazing', 'stop lies', 'pitiful nerd', 'excellent work', 'supreme quality', 'bad', 'highly respectable']
y_train = [1, 0, 0, 1, 1, 0, 1]
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
vocab_size = len(tokenizer.word_index) + 1
print(vocab_size)
```

```python
X_encoded = tokenizer.texts_to_sequences(sentences)
print(X_encoded)
```

```python
max_len = max(len(l) for l in X_encoded)
print(max_len)
```

```python
X_train = pad_sequences(X_encoded, maxlen=max_len, padding='post')
y_train = np.array(y_train)
print(X_train)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, Flatten

embedding_dim = 4

model = Sequential()
model.add(Embedding(vocab_size, embedding_dim, input_length=max_len))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
```

이 셀은 현재 자료가 위키독스 '딥 러닝을 이용한 자연어 처리 입문' 튜토리얼의 일부임을 알려주는 주석입니다. 어떤 자료를 참고하고 있는지 명시하고 있습니다.

이 셀은 텍스트 데이터를 처리하고 딥러닝 모델을 만들기 위해 필요한 라이브러리들을 불러옵니다:
- `numpy`: 숫자 배열을 다루는 데 사용됩니다.
- `Tokenizer`: 텍스트를 숫자로 변환하는 데 사용됩니다.
- `pad_sequences`: 길이가 다른 텍스트 시퀀스들을 일정한 길이로 맞춰주는 데 사용됩니다.

이 셀에서는 분석할 문장들(`sentences`)과 각 문장의 긍정/부정(1/0)을 나타내는 라벨(`y_train`)을 정의합니다. 예를 들어, 'nice great best amazing'은 긍정(1)이고, 'stop lies'는 부정(0)입니다.

이 셀은 `Tokenizer` 객체를 생성하고, 정의된 문장들(`sentences`)을 학습시킵니다. 학습을 통해 각 단어에 고유한 숫자를 부여하고(`tokenizer.fit_on_texts`), 전체 단어의 개수(`vocab_size`)를 계산합니다. `vocab_size`는 임베딩 레이어의 크기를 결정하는 데 사용됩니다.

이 셀은 학습된 `Tokenizer`를 사용하여 `sentences`의 각 단어를 숫자로 변환합니다. 그 결과 각 문장은 숫자의 시퀀스(`X_encoded`)로 표현됩니다. 예를 들어, 'nice great best amazing'은 `[1, 2, 3, 4]`와 같이 변환됩니다.

이 셀은 `X_encoded`에 있는 숫자 시퀀스들 중에서 가장 긴 시퀀스의 길이를 찾습니다. 이 `max_len` 값은 모든 시퀀스의 길이를 통일시킬 때 기준이 됩니다.

이 셀은 `pad_sequences` 함수를 사용하여 `X_encoded`의 모든 시퀀스 길이를 `max_len`에 맞춰줍니다. 짧은 시퀀스에는 0을 뒤에 추가하여 길이를 맞춥니다 (`padding='post'`). 이렇게 변환된 데이터를 `X_train`에 저장하고, `y_train`은 NumPy 배열로 변환합니다.

이 셀은 Keras를 사용하여 딥러닝 모델을 구축합니다:
- `Sequential()`: 순차적인 레이어 스택으로 구성된 모델을 만듭니다.
- `Embedding(vocab_size, embedding_dim, input_length=max_len)`: 단어를 저차원의 밀집 벡터(임베딩 벡터)로 변환하는 레이어입니다. `vocab_size`는 전체 단어의 개수, `embedding_dim`은 임베딩 벡터의 차원, `input_length`는 입력 시퀀스의 최대 길이입니다.
- `Flatten()`: 임베딩 레이어의 출력을 1차원 배열로 평탄화하여 다음 레이어에 전달합니다.
- `Dense(1, activation='sigmoid')`: 출력 레이어입니다. 이진 분류(긍정/부정)를 위해 1개의 뉴런과 'sigmoid' 활성화 함수를 사용합니다. 'sigmoid'는 0과 1 사이의 값을 출력하여 확률을 나타냅니다.

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
model.fit(X_train, y_train, epochs=100, verbose=2)
```

2. 사전 훈련된 GloVe 사용하기

```python
from urllib.request import urlretrieve, urlopen
import gzip
import zipfile
```

```python
import requests

url = "https://nlp.stanford.edu/data/glove.6B.50d.txt"
r = requests.get(url)

with open("glove.6B.50d.txt", "wb") as f:
    f.write(r.content)
```

용량, 다운로드 로딩 이슈로 일부만 다운로드...


```python
embedding_dict = dict()

f = open('glove.6B.100d.txt', encoding="utf8")

for line in f:
    word_vector = line.split()
    word = word_vector[0]

    # 100개의 값을 가지는 array로 변환
    word_vector_arr = np.asarray(word_vector[1:], dtype='float32') #숫자로 변환
    embedding_dict[word] = word_vector_arr #딕셔너리에 저장
f.close()

print('%s개의 Embedding vector가 있습니다.' % len(embedding_dict))
```

```python
print(embedding_dict['respectable'])
print(len(embedding_dict['respectable']))
```

len : "respectable" 단어의 벡터 길이가 몇 개냐?

```python
# 단어 집합 크기의 행과 100개의 열을 가지는 행렬 생성, 값은 전부 0으로 채워진다.
embedding_matrix = np.zeros((vocab_size, 100))
np.shape(embedding_matrix)
```

```python
print(tokenizer.word_index.items())
```

```python
tokenizer.word_index['great']
```

```python
print(embedding_dict['great'])
```

```python
for word, index in tokenizer.word_index.items():
  # 단어와 맵핑되는 사전 훈련된 임베딩 벡터값
  vector_value = embedding_dict.get(word)
  if vector_value is not None:
    embedding_matrix[index] = vector_value
```

```python
embedding_matrix[2]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, Flatten

model = Sequential()
e = Embedding(vocab_size, 100, weights=[embedding_matrix], input_length=max_len, trainable=False)
```

```python
model.add(e)
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
model.fit(X_train, y_train, epochs=100, verbose=2)
```

3. 사전 훈련된 Word2Vec 사용하기

```python
!pip install gensim
import gensim
```

```python
import gensim.downloader as api

word2vec_model = api.load("glove-wiki-gigaword-100")
```

```python
print(word2vec_model['king'][:5])
print(word2vec_model.most_similar('king'))
```

```python
print(word2vec_model.vectors.shape) # 모델의 크기 확인
```

```python
# 단어집합 크기의 행과 100개의 열을 가지는 행렬 생성, 값은 전부 0으로 채워진다
embedding_matrix = np.zeros((vocab_size, 100))
np.shape(embedding_matrix)
```

```python
def get_vector(word):
  if word in word2vec_model:
    return word2vec_model[word]

  else:
    return None
```

```python
for word, index in tokenizer.word_index.items():
  # 단어와 맵핑되는 사전 훈련된 임베딩 벡터값
  vector_value = get_vector(word)
  if vector_value is not None:
    embedding_matrix[index] = vector_value
```

```python
print(word2vec_model['nice'])
```

```python
print(word2vec_model['great'])
```

```python
print(embedding_matrix[2])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, Flatten, Input

model = Sequential()
model.add(Input(shape=(max_len,), dtype='int32'))
e = Embedding(vocab_size, 100, weights=[embedding_matrix], trainable=False)
model.add(e)
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
model.fit(X_train, y_train, epochs=100, verbose=2)
```
