```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. 활성화 함수 정의 (astype으로  구현)
def step_function(x):
    return (x > 0).astype(int)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# 2. 행렬곱을 이용한 신경망 연산 (입력-은닉-출력)
def run_neural_network():
    # 입력 데이터: 2인분 (배고픔, 냉장고, 귀찮음)
    X = np.array([[8, 9, 9],
                  [2, 3, 1]])

    # 가중치와 편향
    W = np.array([0.2, 0.3, 0.6])
    B = -5.0

    # 행렬곱으로 한꺼번에 계산 (병렬 연산)
    tmp = np.dot(X, W) + B

    # 결과 확인
    print("--- 신경망 연산 결과 ---")
    print(f"종합 점수(tmp): {tmp}")
    print(f"외식 여부(Step): {step_function(tmp)}")
    print(f"외식 확률(Sigmoid): {np.round(sigmoid(tmp) * 100, 2)}%")

# 실행
run_neural_network()
```

```python
# -----------------------------------------
# 2. 딥러닝 3층 구조
# -----------------------------------------

# [Step 1] 입력층(Input):2개의 데이터
X = np.array([1.0, 0.5])

# [Step 2] 은닉층 (Hidden Layer): 3개의 뉴런이 복잡하게 계산
W1 = np.array([
    [0.1, 0.3, 0.5],
    [0.2, 0.4, 0.6]
])
B1 = np.array([0.1, 0.2, 0.3])

# 행렬곱 + 편향 더하기
A1 = np.dot(X, W1) + B1

# 활성화 점수 (시그모이드) 통과 -> 다음층으로 전달할 신호 생성
Z1 = sigmoid(A1)

# [Step 3] 출력층 (Output Layer): 최종 2개의 결론 도출
W2 = np.array([
    [0.1, 0.4],
    [0.2, 0.5],
    [0.3, 0.6]
])
B2 = np.array([0.1, 0.2])


# 은닉층 신호(Z1)를 받아서 다시 행렬곱
A2 = np.dot(Z1, W2) + B2


# 최종 결과에 계단 함수 적용
# 0보다 크면 1, 아니면 0으로 잘라 결론을 냅니다
final_decision = step_function(A2)

print("1. 은닉층 통화 신호 (Z1):", Z1)
print("2. 최종 계산 점수 (A2):", A2)
print("3. 최종 합격 / 불합격 결정 (Step):", final_decision)
```

층을 거듭하며 생각: 입력 데이터가 바로 결과가 되는 게 아니라, W1이라는 필터를 거쳐 '은닉층'에서 한 번 가공됩니다. 우리 뇌가 정보를 한 번 더 꼬아서 생각하는 것과 같습니다.

행렬의 크기 변화: 입력은 2개였지만 중간에 뉴런을 3개(W1의 열 개수)로 늘렸다가, 마지막에 결론 2개(W2의 열 개수)로 줄어듭니다. 이 유연한 크기 조절이 행렬곱의 핵심입니다.

활성화 함수의 조합: 중간에는 부드러운 시그모이드를 써서 정보를 섬세하게 전달하고, 맨 마지막에는 계단 함수(astype)를 써서 "최종 합격(1)이냐 불합격(0)이냐"를 딱 잘라 말해주는 구조입니다.
